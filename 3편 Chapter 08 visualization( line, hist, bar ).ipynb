```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df1=pd.DataFrame({'X':[10,24,25,40,55],'Y':[150,160,170,180,177]})

fig,ax=plt.subplots(2,2)

ax[0,0].plot(df1['X'],df1['Y'],color='blue',linestyle='solid',marker='o')
ax[0,1].plot(df1['X'],df1['Y'],color='blue',linestyle='dotted',marker='o')
ax[1,0].plot(df1['X'],df1['Y'],color='blue',linestyle='dashed',marker='o')
ax[1,1].plot(df1['X'],df1['Y'],color='blue',linestyle='dashdot',marker='o')

ax[0, 0].set_title("solid")
ax[0, 1].set_title("dotted")
ax[1, 0].set_title("dashed")
ax[1, 1].set_title("dashdot")

plt.subplots_adjust(wspace=0.2, hspace=0.4)

plt.show()
```

## 제2절 히스토그램(histogram)
### 1. plt.hist(객체, 구간)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### (1) 기본형

```python
x = np.random.normal(size = 1000)  

plt.hist(x) 
plt.title('Seoul') 
plt.legend(['items']) 
plt.show()
```

#### (2) seaborn을 활용한 데이터 출력

```python
sns.histplot(x)
plt.title('Seoul by seaborn')
plt.legend(['items'])
plt.show()
```

#### (3) subplot을 이요한 다수 그래프 출력법

```python
height = np.random.normal(size = 1000)

plt.subplot(3,1,1)
plt.hist(height, bins = 4)

plt.subplot(3,1,2)
plt.hist(height, bins = 'auto')

plt.subplot(3,1,3)
plt.hist(height, bins = 'auto', rwidth = 0.8)
```

## 제3절 막대 그래프 (Bar chart)

#### (1) 기본형

```python
x = ['item1', 'item2', 'item3', 'item4'] 
y = [32, 123, 53, 11]

plt.bar(x,y)
plt.title('seoul univ')
plt.xlabel('items')
plt.ylabel('number of itmes')
plt.legend(['items'])
plt.show()

sns.barplot(x=x,y=y)
plt.title('seoul univ by seaborn')
plt.xlabel('items')
plt.ylabel('number of itmes')
plt.legend(['items'])
plt.show()
```

#### (2) 가로로 막대 그래프 그리기

```python
x = ['item1', 'item2', 'item3', 'item4'] 
y = [32, 123, 53, 11]

plt.barh(x,y)
plt.title('seoul univ')
plt.xlabel('items')
plt.ylabel('number of itmes')
plt.legend(['items'])
plt.show()

sns.barplot(x=x,y=y, orient = 'h')
plt.title('seoul univ by seaborn')
plt.xlabel('items')
plt.ylabel('number of itmes')
plt.legend(['items'])
plt.show()
```
